## Grab profile information of registered athletes in ITTF system from ITTF official website.

```python
ply_profile_df = get_ittf_info.grab_player_data()
```

```python
ply_profile_df
```

```python
ply_profile_df.to_csv('ply_profile_df.csv')
```

## Grab all previous players participation information in World Table Tennis Championships.

```python
tournament_ply_info_df = get_ittf_info.grab_tournament_data()
```

```python
tournament_ply_info_df
```

```python
tournament_ply_info_df.to_csv('tournament_ply_info_df.csv')
```

## Merge dataframes

#### Merge player info to ranking dataset: (currently only focus on male athletes)

```python
ply_rank_and_profile_df = male_ply_rank_info.merge(ply_profile_df, left_on='id', right_on='id')

# clean dataset:
ply_rank_and_profile_df = ply_rank_and_profile_df.drop(['full_name_x','YOB_y','rank_increment','point_increment'],axis=1)

#rename column name:
ply_rank_and_profile_df.columns =['rank','YOB','country','rating','id','full_name','assoc','gender','activity','playing_hand','playing_style','grip']
```

```python
ply_rank_and_profile_df.head(10)
```

```python
# Left-handed penhold athletes who are within 1000 ranking:
ply_rank_and_profile_df.loc[(ply_rank_and_profile_df['playing_hand'] == 'Left-handed')& (ply_rank_and_profile_df['grip'] == 'Penhold'),]

```

```python
# Right-handed penhold athletes who are within 1000 ranking:
ply_rank_and_profile_df.loc[(ply_rank_and_profile_df['playing_hand'] == 'Right-handed')& (ply_rank_and_profile_df['grip'] == 'Penhold'),]
```

### merge player profile to tournament dataset:

```python
tournament_ply_rank_profile_df = tournament_ply_info_df.merge(ply_profile_df[['id', 'full_name', 'YOB', 'activity', 'playing_hand', 'playing_style', 'grip']], left_on='full_name', right_on='full_name')
```

```python
#show Chinese athletes who attends 2019 World Table Tennis Championships in Budapest

tournament_ply_rank_profile_df.loc[(tournament_ply_rank_profile_df['player_country'] == 'CHN') & (tournament_ply_rank_profile_df['year'] == '2019'),]
```

## Presentation

### Bar Chart
Try to visualize top number of players who attended World Table Tennis Championships most of the times

```python
num_of_display = 10
get_ittf_info.vis_frequency_bar_chart(tournament_ply_info_df,'full_name','full name','frequency','Top %s Players who attended World Table Tennis Championships most of the times'%num_of_display,num_of_display)
```

Try to visualize top number of associations which have most number of registered athletes

```python
num_of_display = 10
get_ittf_info.vis_frequency_bar_chart(ply_profile_df,'assoc','assoc','frequency','Top %s associations which have most number of registered athletes'%num_of_display,num_of_display)
```

We can see USA has most number of ITTF registered table tennis athletes, followed by CHN and RUS

Try to visualize top number of WTTC events which have most number of participants

```python
# which year's tournament has most participants

interest_df = tournament_ply_info_df[['year','venue_city']].copy()
interest_df['year_city'] = interest_df.agg('_'.join, axis=1)
```

```python
num_of_display = 30
get_ittf_info.vis_frequency_bar_chart(interest_df,'year_city','year_city','frequency','Top %s WTTC with largest number of participants'%num_of_display,num_of_display)


```

We can see that WTTC in Manchester in 1997 had largest number of participants
